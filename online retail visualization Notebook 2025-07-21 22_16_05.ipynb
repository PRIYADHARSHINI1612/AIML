```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the online source
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv"
online_rt = pd.read_csv(url, encoding='latin1')

# Keep only positive quantities
online_rt = online_rt[online_rt['Quantity'] > 0]

# Group by country and sum quantities
countries = online_rt.groupby('Country')['Quantity'].sum().sort_values(ascending=False)

# Exclude United Kingdom and select top 10
top_countries = countries.drop('United Kingdom').head(10)

# Create bar plot
top_countries.plot(kind='bar', color='skyblue', edgecolor='black')

# Set labels and title
plt.xlabel('Country')
plt.ylabel('Total Quantity Ordered')
plt.title('Top 10 Countries by Quantity (Excluding UK)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
online_rt = online_rt[online_rt.Quantity > 0]
online_rt.head()
```

```python
customers = online_rt.groupby(['CustomerID','Country']).sum()
customers = customers[customers.UnitPrice > 0]
customers['Country'] = customers.index.get_level_values(1)
top_countries = ['Netherlands', 'EIRE', 'Germany']
customers = customers[customers['Country'].isin(top_countries)]
```

```python
g =sns.FacetGrid(customers, col="Country", col_wrap=3, height=4, sharex=False, sharey=False)

# Map scatterplot
g.map(plt.scatter, "Quantity", "UnitPrice", alpha=0.5)

# Add legend (though scatter plots don't have a legend by default unless using hue)
g.add_legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sales_volume =online_rt.groupby('Country').Quantity.sum().sort_values(ascending=False)
top3 = sales_volume.index[1:4] #We are excluding UK
top3
index = pd.Index(['Netherlands', 'EIRE', 'Germany'], dtype='object',
name='Country')
```

```python
online_rt['Revenue'] = online_rt.Quantity * online_rt.UnitPrice
online_rt.head()
```

```python
grouped=online_rt[online_rt.Country.isin(top3)].groupby(['CustomerID','Country'])
plottable = grouped['Quantity','Revenue'].agg('sum')
plottable['AvgPrice'] = plottable.Revenue / plottable.Quantity
# get the value of the index and put in the column Country
plottable['Country'] = plottable.index.get_level_values(1)
plottable.head()
```

```python
g = sns.FacetGrid(plottable, col="Country")
g.map(plt.scatter, "Quantity", "AvgPrice", alpha=1)
g.add_legend();
```

```python
price_start = 0
price_end = 50
price_interval = 1
#Creating the buckets to collect the data accordingly
buckets = np.arange(price_start,price_end,price_interval)
#Select the data and sum
revenue_per_price = online_rt.groupby(pd.cut(online_rt.UnitPrice,buckets)).Revenue.sum()
revenue_per_price.head() 
```

```python
revenue_per_price.plot()
plt.xlabel('Unit Price (in intervals of '+str(price_interval)+')')
plt.ylabel('Revenue')
plt.show()
```

```python
revenue_per_price.plot()
plt.xlabel('Unit Price (in buckets of '+str(price_interval)+')') 
plt.ylabel('Revenue')
plt.xticks(np.arange(price_start,price_end,3), np.arange(price_start,price_end,3))
plt.yticks([0, 500000, 1000000, 1500000, 2000000, 2500000],['0', '$0.5M', '$1M', '$1.5M', '$2M', '$2.5M'])
plt.show()
```
